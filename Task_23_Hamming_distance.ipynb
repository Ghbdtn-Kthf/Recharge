1. What is the maximum Hamming distance between two data rows in the file?
Hamming Distance is the number of places where two vectors differ.

```python
import itertools
from scipy.spatial.distance import hamming

ids = dataset["ID"].values
```

```python
maximum = 0
for i in itertools.combinations(ids, 2):
    a = list((dataset[dataset["ID"] == i[0]]).values)[0]
    b = list((dataset[dataset["ID"] == i[1]]).values)[0]
    if  (hamming(a, b) * len(a)) > maximum:
      maximum = (hamming(a, b) * len(a))
print(maximum)
```
